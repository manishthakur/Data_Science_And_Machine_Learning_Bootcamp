```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 - A simple  cost function
## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x +1
```

```python
# Make data

x_1 = np.linspace(start=-3, stop=3, num=10)
x_1
```

```python
plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 8])
plt.ylabel('Y', fontsize=16)
plt.xlabel('X', fontsize=16)
plt.show()
```
