# Example 1 - A simple  cost function
## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x +1
```

```python
# Make data

x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 8])
plt.ylabel('f(x)', fontsize=16)
plt.xlabel('X', fontsize=16)
plt.show()
```

## Slope & Derivatives
### Create a function for derivates of f(x) called df(x)

```python
def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side

plt.figure(figsize=[15, 5])

#1 Chart : Cost function
plt.subplot(1, 2, 1)
plt.title('Cost function', fontsize=17)

plt.plot(x_1, f(x_1),color='blue',  linewidth=5)
plt.xlim([-3, 3])
plt.ylim([0, 8])

plt.ylabel('f(x)', fontsize=16)
plt.xlabel('X', fontsize=16)

#2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.ylabel('df(x)', fontsize=16)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.show()
```

```python
# Gradient decent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]


for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    # print(step_size)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('Loop ran this many time:', n)
        break
    
print('Local minimum occures at:', new_x)
print('Slope or df(x) value at this point is: ', df(new_x))
print('f(x) vakue of cost at this point is:', f(new_x))
    
```

```python
# Plot function and derivative side by side

plt.figure(figsize=[15, 5])

#1 Chart : Cost function
plt.subplot(1, 3, 1)

plt.xlim([-3, 3])
plt.ylim([0, 8])

plt.title('Cost function', fontsize=17)

plt.ylabel('f(x)', fontsize=16)
plt.xlabel('X', fontsize=16)

plt.plot(x_1, f(x_1),color='blue',  linewidth=3)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', alpha=0.6)

#2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.ylabel('df(x)', fontsize=16)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

#3 Chart: Derivative Close Up
plt.subplot(1, 3, 3)

plt.title('Gradient Decent (close up)', fontsize=17)
plt.ylabel('df(x)', fontsize=16)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.5)

plt.show()
```

## Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
            
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print('Local min occurs at:', local_min)
print('Number of steps: ', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.5, 
                                                 multiplier=0.02, precision=0.001)
print('Local min occurs at:', local_min)
print('Number of steps: ', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.1)
print('Local min occurs at:', local_min)
print('Number of steps: ', len(list_x))
```
